```python
list8=[3,1,4]
list9=[2,5,0]
combo=list8+list9
for i in range(len(combo)):
    for j in range(0,len(combo)-1):
        if combo[j]>combo[j+1]:
            combo[j],combo[j+1]=combo[j+1],combo[j]
print("After swapping")
print(combo)
    
```

```python
nested_list=[1,[2,3],4,[5,6,7],8]
list1=[]
for i in nested_list:
    if type(i)==list:
        for j in i:
            list1.append(j)
    else:
        list1.append(i)
print(list1)
```
