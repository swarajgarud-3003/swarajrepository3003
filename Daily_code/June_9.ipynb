# inheritance

```python
class Person:
    insti="MITU"
    def __init__(self,age,ht):
        self.age=age
        self.ht=ht
    def show(self):
        print(f"age->{self.age},height->{self.ht}.institue->{self.insti}")
```

```python
s1=Person(18,5.6)
```

```python
s1.show()
```

```python
class ravi(Person):
    def __init__(self,age,ht,sal):
        super().__init__(age,ht)
        self.sal=sal
    def show_sal(self):
        print(f"salary:{self.sal}")
    
```

```python
p1=ravi(18,5.9,500)
```

```python
p1.show_sal()
```

```python
p1.show()
```

```python
class Menuapp:
    def __init__(self):
        self.itemlist=[]
    def show_menu(self):
        print("Menu\n 1.add item \n2.view item \n3.exit")
    def add_items(self):
        item=input("enter the item")
        self.itemlist.append(item)
    def view_item(self):
        return self.itemlist
    def run(self):
        while True:
            self.show_menu()
            ch=int(input("enter the choice"))
            if ch==1:
                self.add_items()
            elif ch==2:
                self.view_items()
            elif ch==3:
                break
            else:
                print("invalid choice")
```

```python
m=Menuapp()
```
