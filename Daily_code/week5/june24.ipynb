```python
def remove_outliers(ads, column):
    Q1 = ads[column].quantile(0.25)
    Q3 = ads[column].quantile(0.75)

    IQR = Q3 - Q1

    lowerb = Q1 - 1.5 * IQR
    upperb = Q3 + 1.5 * IQR

    print("Q1 =", Q1)
    print("Q3 =", Q3)
    print("IQR =", IQR)
    print("Lower Bound =", lowerb)
    print("Upper Bound =", upperb)

    ads_clean = ads[(ads[column] >= lowerb) & (ads[column] <= upperb)]

    return ads_clean
```

```python
ads=remove_outliers(ads,"Age")
```
