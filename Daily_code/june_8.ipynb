```python
class  Person:
    def __init__(self,name,age,ht):
        self.name=name
        self.__age=age # use __ for private variable
        self.height=ht
    def show_info(self):
        return f"name -> {self.name}, age -> {self.__age}, height -> {self.height}"
        
        
        

```

```python
p1=Person("swaraj",20,5.8)
p2=Person("sachin",22,5.9)
```

```python
p1.show_info()
```

```python
"""
create a class called  employee =>
attributees -> company,name,monthly_salary,job_designation
here all the objects of this company will have the same common company make salary as private variable
create 3 objects -> e1,e2,e3 with given attributes

method-> 1.method to calculate the anual salary
         2.method to show entire data
""";
```

```python
class  Employee:
    def __init__(self,company_name,employee_name,monthly_salary,job_designation):
        self.companyname=company_name
        self.employeename=employee_name
        self.__monthlysalary=monthly_salary
        self.jobdesignation=job_designation
    def calculate_anualsalary(self):
        return f"anual_salary,{self.__monthlysalary*12}"
    def show_data(self):
        return f"company name:{self.companyname},employee name{self.employeename},monthly salary{self.__monthlysalary},job designation{self.jobdesignation}"
        
        
        
```

```python
e1=Employee("abc","swaraj",25000,"hr")
e2=Employee("abc","suraj",25000,"xyz")

e3=Employee("abc","aryan",25000,"axy")
```

```python
e1.show_data()
```
