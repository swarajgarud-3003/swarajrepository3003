```python
try:
    f= open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("file does not exist")
except:
    print("something went wrong")
```

```python
try:
    f=open("a1.txt")
    print(var1)
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)
```

```python
try:
    f=open("a1.txt")
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
```

```python
# finally
```

```python
try:
    f=open("a1.txt")
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
finally:
    print("run completed")
```

```python
try:
    f=open("a1.txt")
    if f.name=="a1.txt":
        raise Exception
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
finally:
    print("run completed")
```

```python
try:
    n1=int(input("enter the number"))
    n2=int(input("enter the number"))
    print(n1%n2)
except ZeroDivisionError:
    print("number does not divide by zero")
except ValueError:
    print("different value")
    
```
