Accept a temperature in celsius from the user,convert it to fahrenheit and kelvin,and print all three values with labels.

```python
celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (celsius * 9/5) + 32
kelvin = celsius + 273.15
print("Temperature in Celsius:", celsius)
print("Temperature in Fahrenheit:", fahrenheit)
print("Temperature in Kelvin:", kelvin)
```

Take a year as input and print whether it is a leap year or not.



```python
year = int(input("Enter a year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

Take a positive integer n as input and print all prime numbers from 2 to n.



```python
n =int(input("enter the number"))

for i in range(2, n + 1):
    prime = True

    for j in range(2, i):
        if i % j == 0:
            prime = False
            break

    if prime:
        print(i)
```

9. Keep accepting numbers from the user until they enter 0. Then print the total count of numbers entered, their sum, and their average. Do not include 0 in the calculations.

```python
count = 0
total = 0

while True:
    num = int(input("Enter a number (0 to stop): "))

    if num == 0:
        break

    count += 1
    total += num

if count > 0:
    average = total / count
    print("Count =", count)
    print("Sum =", total)
    print("Average =", average)
else:
    print("No numbers entered")
```

Take n as input and print a triangle pattern of n rows made of asterisks,where row i has(2i -1)stars centered appropriately

```python
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    spaces = n - i
    stars = 2 * i - 1
    print(" " * spaces + "*" * stars)
```
