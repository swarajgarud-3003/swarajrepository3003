Write a function analyze_sentence(sentence) that takes a sentence and returns a dictionary with word_count, char_count (excluding spaces), unique_words (case-insensitive), and most_frequent_word. Test it with:

"The quick brown fox jumps over the lazy dog the fox"



```python
def analyze_sentence(sentence):
    words = sentence.lower().split()

    word_count = len(words)
    char_count = len(sentence.replace(" ", ""))
    unique_words = len(set(words))

    freq = {}
    for word in words:
        freq[word] = freq.get(word, 0) + 1

    most_frequent_word = max(freq, key=freq.get)

    return {
        "word_count": word_count,
        "char_count": char_count,
        "unique_words": unique_words,
        "most_frequent_word": most_frequent_word
    }

sentence = "The quick brown fox jumps over the lazy dog the fox"
print(analyze_sentence(sentence))


```

Write a function format_report(data) that takes a list of tuples (name, score, grade) and prints a formatted table where name is left-aligned (width 15), score is right-aligned (width 6), and grade is centered (width 8). Use this data:data = [("Alice", 92, "A"), ("Bob", 78, "B+"), ("Charlie", 55, "C"), ("Diana", 100, "A+")]

```python
def format_report(data):
    print(f"{'Name':<15}{'Score':>6}{'Grade':^8}")
    print("-" * 29)

    for name, score, grade in data:
        print(f"{name:<15}{score:>6}{grade:^8}")

data = [
    ("Alice", 92, "A"),
    ("Bob", 78, "B+"),
    ("Charlie", 55, "C"),
    ("Diana", 100, "A+")
]

format_report(data)
```

 Write a function inventory_manager(records) that takes a list of tuples (product, category, quantity, price) and returns a dictionary grouped by category. Each category should have total_items, total_value, and a list of product names. Use:

records = [
    ("Laptop", "Electronics", 10, 55000),
    ("Mouse", "Electronics", 50, 499),
    ("Desk", "Furniture", 8, 12000),
    ("Chair", "Furniture", 15, 4500),
    ("Pen", "Stationery", 200, 20),
]

Section C — Functions and Modules



```python
def inventory_manager(records):
    inventory = {}

    for product, category, quantity, price in records:

        if category not in inventory:
            inventory[category] = {
                "total_items": 0,
                "total_value": 0,
                "products": []
            }

        inventory[category]["total_items"] += quantity
        inventory[category]["total_value"] += quantity * price
        inventory[category]["products"].append(product)

    return inventory


records = [
    ("Laptop", "Electronics", 10, 55000),
    ("Mouse", "Electronics", 50, 499),
    ("Desk", "Furniture", 8, 12000),
    ("Chair", "Furniture", 15, 4500),
    ("Pen", "Stationery", 200, 20),
]

result = inventory_manager(records)

for category, details in result.items():
    print(category, ":", details)
```

a) Write a closure make_multiplier(n) that returns a function multiplying its input by n. Create triple and quintuple using it.

```python

def make_multiplier(n):
    def multiply(x):
        return x * n
    return multiply

triple = make_multiplier(3)
quintuple = make_multiplier(5)

print("Triple of 4 =", triple(4))
print("Quintuple of 4 =", quintuple(4))
```

Write a decorator @timer that prints the execution time of any function it wraps. Apply it to slow_sum(n) which computes the sum of squares from 1 to n.

```python
import time

def timer(func):
    def wrapper(n):
        start = time.time()
        result = func(n)
        end = time.time()

        print("Execution Time:", end - start, "seconds")
        return result
    return wrapper

@timer
def slow_sum(n):
    total = 0
    for i in range(1, n + 1):
        total += i ** 2
    return total

print("Sum =", slow_sum(100000))
```

c) Write a decorator @validate_positive that raises a ValueError if any argument is zero or negative. Apply it to safe_sqrt(x).

```python
import math

def validate_positive(func):
    def wrapper(x):
        if x <= 0:
            raise ValueError("Value must be positive")
        return func(x)
    return wrapper

@validate_positive
def safe_sqrt(x):
    return math.sqrt(x)

print(safe_sqrt(25))

# print(safe_sqrt(-4))  # Raises ValueError
```
