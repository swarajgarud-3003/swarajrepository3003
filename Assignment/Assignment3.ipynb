1. Create a class BankAccount with the following:

a) Attributes: account_holder, account_number, and balance.

b) Methods: deposit(amount), withdraw(amount), and get_balance().

c) withdraw() should print a message and do nothing if the amount exceeds the current balance.

d) A __str__ method that returns a readable summary of the account.

Create two BankAccount objects and perform a few deposits and withdrawals on each.

```python
class BankAccount:
    def __init__(self, account_holder, account_number, balance):
        self.account_holder=account_holder
        self.account_number=account_number
        self.balance=balance

    def deposit(self,amount):
        self.balance=self.balance+ amount
        print(f" deposit balance:{self.balance}")

    def withdraw(self,withdraw_amount):
        if withdraw_amount > self.balance:
            print(f"withdrawal not allowed")
        else:
            self.balance=self.balance-withdraw_amount
            print(f"withdraw:{self.balance}")
    def get_balance(self):
        print(f"balance{self.balance}")

    def __str__(self):
        print(f"Account holder{self.account_holder},Account_number{self.account_number},Account balance{self.balance}")
```

3. Create a class Employee with private attributes __name, __emp_id, and __salary. Add:

a) Getter and setter methods for all three attributes.

b) The salary setter should raise a ValueError if the value is below 10000.

c) A method give_raise(percent) that increases salary by the given percentage.

Demonstrate that the private attributes cannot be accessed directly, and that all access goes through the methods.

```python
class Employee:
    def __init__(self,name,emp_id,salary):
        self.__name=name
        self.__emp_id=emp_id
        self.__salary=salary
    def get_name(self):
        return self.__name
    def get_emp_id(self):
        return self.__emp_id
    def get_salary(self):
        return self.__salary

    def set_name(self,name):
        self.__name=name
    def set_emp_id(self,emp_id):
        self.__emp_id=emp_id
    def set_salary(self,salary):
        if salary < 10000:
            raise ValueError("salary must be at least 10000")
        self.__salary=salary
        return self.__salary

    def give_raise(self,percent):
        self.percent=self.__salary+(self.__salary*percent/100)
        return self.percent
        
```

b) A method get_entry(pin, key) that returns the value only if the pin matches.

c) Any wrong pin attempt should print 'Access Denied' and return nothing.

```python
class SecureVault:
    def __init__(self,pin):
        self.__pin=pin
        self.__data={}
    def add_entry(self,pin,key,value):
        if pin == self.__pin:
            self.__data[key]=value
            print("enter added successful")
        else:
            print("access denied")
            return
    def get_entry(self,pin,key):
        if pin ==self.__pin:
            return self.__data.get(key)
        else:
            print("acces denied")
            return
        



        
```

Section F — Menu-Driven App with OOP
10.

Build a menu-driven Student Record Manager using OOP and file handling. The app should have a class StudentManager with the following menu options:

Add Student — takes name, roll number, and marks for 3 subjects.
View All — displays all stored students with their average marks.
Search Student — finds a student by roll number.
Delete Student — removes a student by roll number.
Save to File — writes all records to students_data.txt.
Load from File — reads records back from students_data.txt on startup.
Exit

Requirements: use a class with a list as internal storage, handle invalid menu choices gracefully, and use a while loop to keep the menu running until the user chooses to exit.

```python
class StudentManager:
    def __init__(self):
        self.students = []

    def add_student(self):
        name = input("Enter Name: ")
        roll = input("Enter Roll Number: ")
        m1 = float(input("Enter Marks of Subject 1: "))
        m2 = float(input("Enter Marks of Subject 2: "))
        m3 = float(input("Enter Marks of Subject 3: "))
        self.students.append([name, roll, m1, m2, m3])

    def view_all(self):
        if len(self.students) == 0:
            print("No record found")
        else:
            for student in self.students:
                avg = (student[2] + student[3] + student[4]) / 3
                print("Name:", student[0])
                print("Roll Number:", student[1])
                print("Average Marks:", avg)
                print()

    def search_student(self):
        roll = input("Enter Roll Number to Search: ")
        for student in self.students:
            if student[1] == roll:
                avg = (student[2] + student[3] + student[4]) / 3
                print("Name:", student[0])
                print("Roll Number:", student[1])
                print("Average Marks:", avg)
                return
        print("Student Not Found!")

    def delete_student(self):
        roll = input("Enter Roll Number to Delete: ")
        for student in self.students:
            if student[1] == roll:
                self.students.remove(student)
                print("Student Deleted Successfully!")
                return
        print("Student Not Found!")

    def save_to_file(self):
        with open("students_data.txt", "w") as file:
            for student in self.students:
                file.write(
                    f"{student[0]},{student[1]},{student[2]},{student[3]},{student[4]}\n"
                )
        print("Data Saved Successfully!")

    def load_from_file(self):
        self.students = []
        try:
            with open("students_data.txt", "r") as file:
                for line in file:
                    data = line.strip().split(",")
                    self.students.append(
                        [
                            data[0],
                            data[1],
                            float(data[2]),
                            float(data[3]),
                            float(data[4]),
                        ]
                    )
        except FileNotFoundError:
            pass


manager = StudentManager()
manager.load_from_file()

while True:
    print("\n1.Add  2.View  3.Search")
    print("4.Delete  5.Save  6.Load  7.Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        manager.add_student()
    elif choice == "2":
        manager.view_all()
    elif choice == "3":
        manager.search_student()
    elif choice == "4":
        manager.delete_student()
    elif choice == "5":
        manager.save_to_file()
    elif choice == "6":
        manager.load_from_file()
    elif choice == "7":
        break
    else:
        print("Invalid Choice")
```
